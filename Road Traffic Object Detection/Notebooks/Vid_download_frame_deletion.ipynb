```python
import subprocess, sys

url = "https://www.youtube.com/watch?v=wqctLW0Hb_0"

subprocess.run([
    sys.executable,
    "-m",
    "yt_dlp",
    "-f", "bv*+ba/b",
    "--merge-output-format", "mp4",
    "-o", "video.%(ext)s",
    url
])
```

```python
import json, subprocess

probe = subprocess.run([
    "ffprobe",
    "-v","quiet",
    "-print_format","json",
    "-show_format",
    "video.mp4"
], capture_output=True, text=True)

duration = float(json.loads(probe.stdout)["format"]["duration"])
print("Duration:", duration)
```

```python
trim_seconds = 10

subprocess.run([
    "ffmpeg",
    "-y",
    "-i","video.mp4",
    "-t", str(duration - trim_seconds),
    "-c","copy",
    "video_trimmed.mp4"
])

print("Saved -> video_trimmed.mp4")
```
